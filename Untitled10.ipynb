```python
# Cell 1: Setup & Imports (no scikit-surprise)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
from google.colab import files

print(" Libraries imported successfully.")
```

```python
# Cell 3: Load and explore the data

# Load the ratings and books data
ratings = pd.read_csv("ratings.csv")
books = pd.read_csv("books.csv")

# Display basic info
print("Ratings dataset shape:", ratings.shape)
print("Books dataset shape:", books.shape)

# Show first few rows
print("\nSample ratings:")
print(ratings.head())

print("\nSample books:")
print(books[['book_id', 'title', 'authors']].head())

# Check for missing values
print("\nMissing values in ratings:", ratings.isnull().sum().sum())
print("Missing values in books:", books.isnull().sum().sum())
```

```python
# Cell 4: Preprocessing and creating the user-item rating matrix

# Step 1: Filter out users and books with very few ratings to reduce sparsity
# Keep users with at least 50 ratings and books with at least 50 ratings

min_user_ratings = 50
min_book_ratings = 50

filtered_users = ratings['user_id'].value_counts()[ratings['user_id'].value_counts() >= min_user_ratings].index
filtered_books = ratings['book_id'].value_counts()[ratings['book_id'].value_counts() >= min_book_ratings].index

ratings_filtered = ratings[ratings['user_id'].isin(filtered_users) & ratings['book_id'].isin(filtered_books)]

print("Original ratings:", len(ratings))
print("Filtered ratings:", len(ratings_filtered))

# Step 2: Create a user-item matrix
user_item_matrix = ratings_filtered.pivot_table(index='user_id', columns='book_id', values='rating')

print("\nUser-item matrix shape:", user_item_matrix.shape)
print("Number of missing ratings:", user_item_matrix.isna().sum().sum())

# Fill missing values with 0 for clustering
user_item_matrix_filled = user_item_matrix.fillna(0)

print("Matrix ready for clustering.")
```

```python
# Cell 5: Apply K-Means clustering to group similar users

# Scale the matrix for better clustering performance
scaler = StandardScaler()
user_matrix_scaled = scaler.fit_transform(user_item_matrix_filled)

# Choose the number of clusters (you can experiment with this)
k = 5  # start with 5 clusters

kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
user_clusters = kmeans.fit_predict(user_matrix_scaled)

# Add cluster labels back to users
user_cluster_df = pd.DataFrame({
    'user_id': user_item_matrix_filled.index,
    'cluster': user_clusters
})

print("Clustering completed.")
print(user_cluster_df['cluster'].value_counts())

# Visualize the cluster distribution
plt.figure(figsize=(6,4))
sns.countplot(x='cluster', data=user_cluster_df, palette='viridis')
plt.title("User Cluster Distribution")
plt.xlabel("Cluster")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Cell 6: Collaborative Filtering within clusters

# Merge cluster info with the ratings data
ratings_clustered = ratings_filtered.merge(user_cluster_df, on='user_id')

def get_similar_users_in_cluster(target_user_id, cluster_id, top_n=5):
    """Return top N similar users within the same cluster using cosine similarity."""
    # Subset matrix for the target cluster
    cluster_users = user_cluster_df[user_cluster_df['cluster'] == cluster_id]['user_id']
    cluster_matrix = user_item_matrix_filled.loc[cluster_users]

    # Compute cosine similarity between users
    similarity_matrix = cosine_similarity(cluster_matrix)
    similarity_df = pd.DataFrame(similarity_matrix, index=cluster_users, columns=cluster_users)

    # Get top similar users for the target user
    if target_user_id not in similarity_df.index:
        return []
    similar_users = similarity_df[target_user_id].sort_values(ascending=False)[1:top_n+1]
    return similar_users

# Example test: pick one user and get similar users within their cluster
example_user = user_cluster_df['user_id'].sample(1).values[0]
example_cluster = user_cluster_df.loc[user_cluster_df['user_id'] == example_user, 'cluster'].values[0]

print(f"Example user: {example_user}, in cluster {example_cluster}")
similar_users = get_similar_users_in_cluster(example_user, example_cluster)
print("\nTop similar users within the same cluster:")
print(similar_users)
```

```python

# Cell 7: Hybrid recommendation function (manual user input)

def hybrid_recommend(user_id, top_n=5):
    # Find which cluster the user belongs to
    if user_id not in user_cluster_df['user_id'].values:
        print("User not found in clustered dataset.")
        return pd.DataFrame()
    cluster_id = user_cluster_df.loc[user_cluster_df['user_id'] == user_id, 'cluster'].values[0]

    # Get similar users in that cluster
    similar_users = get_similar_users_in_cluster(user_id, cluster_id, top_n=10)
    if len(similar_users) == 0:
        print("No similar users found.")
        return pd.DataFrame()

    # Get ratings of similar users
    similar_user_ids = similar_users.index
    cluster_ratings = ratings_filtered[ratings_filtered['user_id'].isin(similar_user_ids)]

    # Exclude books already rated by target user
    user_rated_books = set(ratings_filtered[ratings_filtered['user_id'] == user_id]['book_id'])
    candidate_books = cluster_ratings[~cluster_ratings['book_id'].isin(user_rated_books)]

    # Compute average rating per book from similar users
    recommended_books = candidate_books.groupby('book_id')['rating'].mean().sort_values(ascending=False).head(top_n)
    recommended_books = recommended_books.reset_index()

    # Merge with book titles
    recommended_books = recommended_books.merge(books[['book_id', 'title', 'authors']], on='book_id')

    return recommended_books[['book_id', 'title', 'authors', 'rating']]

# --- Manual input version ---
user_id_input = int(input("Enter a user ID to get recommendations: "))
top_n_input = 5  # you can change this number if you like

recommendations = hybrid_recommend(user_id_input, top_n=top_n_input)

if not recommendations.empty:
    print(f"\nTop {top_n_input} recommended books for user {user_id_input}:")
    print(recommendations)
else:
    print("No recommendations could be generated for this user.")
```

```python
# Cell 9: Visualization and Summary
# Average rating per cluster
cluster_avg_ratings = (
    ratings_filtered
    .merge(user_cluster_df, on='user_id')
    .groupby('cluster')['rating']
    .mean()
    .reset_index()
)

plt.figure(figsize=(6,4))
sns.barplot(x='cluster', y='rating', data=cluster_avg_ratings, palette='crest')
plt.title("Average Rating by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Average Rating")
plt.show()
print("This graph shows the average rating of users within each cluster. It helps to understand user sentiment differences.")

```

### **Cell 8: Evaluate K-Means Clustering using the Elbow Method**

To determine an optimal number of clusters, `k`, we can use the Elbow Method. This involves calculating the Within-Cluster Sum of Squares (WCSS) for a range of `k` values. The 'elbow' point on the plot, where the rate of decrease in WCSS significantly changes, often indicates a suitable `k`.

```python
wcss = []
max_k = 10 # You can adjust this range

for i in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(user_matrix_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, max_k + 1), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.xticks(range(1, max_k + 1))
plt.grid(True)
plt.show()

print(f"The elbow method helps to visualize the optimal number of clusters (k) by plotting WCSS against k. The 'elbow' point suggests a good trade-off between minimizing WCSS and having a reasonable number of clusters.")
```
